```python
import warnings
warnings.filterwarnings("ignore", category=UserWarning, module="qiskit_ibm_provider")
```

```python
from qiskit import QuantumCircuit
from qiskit import Aer
from qiskit import transpile
from qiskit_ibm_runtime import QiskitRuntimeService, Sampler, Estimator
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit

```

```python
# Building Quantum Circuit & define 16 Quantum bits and 16 classical bits 
q = QuantumRegister(16, 'q')
c = ClassicalRegister(16, 'c')
circuit = QuantumCircuit(q, c)
```

```python
# Apply Hadamard gates to all qubits
circuit.h(q)
# Measure all qubits(breaks the superpositioning of q bit into random classical bits)
circuit.measure(q, c)
circuit.draw()
```

```python
# Execute Quantum Circuit
backend = Aer.get_backend('aer_simulator')

# Run the circuit (no execute(), no transpile)
job = backend.run(circuit, shots=1)

print("Executing Job.....")

# Get the result
result = job.result()
counts = result.get_counts()
print("Result:", counts)
```
